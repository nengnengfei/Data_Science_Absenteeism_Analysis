# Notes

Why drop ID column: ID is the individual identification of each person, and each number from ID column represents a specific employee, indicates precisely who has been away during working hours, ID information could help company management when monitoring the presence of each employee. However, ID information will not improve the analysis as it is an nominal data and it not carrys any numeric information and it will harm the precision of estimations


quantitative analysis: add numeric meaning to categorical nominal values
   * Way to do it: Turn the nominal values into dummy variables ( dummy variable: an explanatory binary variable that equals 1 if a certain categorical effect is present, and that equals 0 if that same effect is absent)
   * Therefore, we want a column where the values of 1 will appear in case an individual has been absent because of reason number 1, and 0 if she was absent because of another reason
